```python
import numpy as np
import pandas as pd
import json
from urllib.request import urlopen
from urllib.parse import quote   
```

```python
apikey = open("APIKEY", "r").read().rstrip()
```

```python
def ch2en(ch_name, apikey):
    url="https://maps.googleapis.com/maps/api/geocode/json?address={}&key={}".format(quote(ch_name), apikey)
    response = urlopen(url)
    jsongeocode = json.loads( response.read().decode('utf-8') )
    return jsongeocode["results"][0]["formatted_address"]
```

# 2018 公投

## 公投投票人人數統計（縣市別）

```python
datapath = "rawdata/referendum_voters_stats.csv"
df_by_city = pd.read_csv(datapath, skipfooter=0, header=None, names=["縣市", "合計", "男", "女"], comment="#")
df_by_city
```

```python
city_names = ["total"]
for i in range(1,len(df_by_city)):
    ch_name = df_by_city.iloc[i,0]
    print(ch_name)
    en_name = ch2en(ch_name, apikey)
    print(en_name)
    city_names.append( en_name.split(",")[0] )
city_names
```

```python
df_by_city["city"] = city_names
```

```python
df_by_city
```

### 匯出清理後的資料

```python
output_filename = "data/referendum_voters_stats.csv"
df_by_city.to_csv(output_filename)
```

## 公投投票人人數統計（年齡別）

```python
datapath = "rawdata/referendum_voters_by_age.csv"
df_by_age_cols = pd.read_csv(datapath, header=None, comment="#", nrows=2)
df_by_age_cols
```

```python
age_groups = df_by_age_cols.iloc[0,2::3]
age_groups.values
```

```python
sex_groups = df_by_age_cols.iloc[1,1::]
sex_groups.values
```

```python
datapath = "rawdata/referendum_voters_by_age.csv"
df_by_age = pd.read_csv(datapath, header=None, comment="#", skiprows=7, index_col=0)
df_by_age.T
```

```python
pairs = [(age, sex) for age in age_groups.values for sex in sex_groups.values[0:3] ]
index = pd.MultiIndex.from_tuples(pairs, names=['年齡', '性別'])
df_by_age2 = df_by_age.T
df_by_age2.index = index
df_by_age3 = df_by_age2.T
df_by_age3
```

### 匯出清理後的資料

```python
output_filename = "data/referendum_voters_stats_by_age.csv"
df_by_age3.to_csv(output_filename)
```

```python
sex_groups = df_by_type_cols.iloc[4,1::]
sex_groups.values
```

```python
df_by_city = df_by_city.T
df_by_city.index = index
df_by_city = df_by_city.T
df_by_city
```

```python
output_filename = "data/regional_voters_stats_by_type.csv"
df_by_type.to_csv(output_filename)
```
